```python
# issue with normalization above, therefore fix here:

dfc = deepcopy(df)

normalized_values = []

for sc in np.unique(dfc["selection coefficient"])[::-1]:
    dfc_sc = deepcopy(dfc[dfc["selection coefficient"] == sc])    
    normvalue = dfc_sc[dfc_sc["1/b"] == 1.0]["fixation time"].mean()        
    normalized_value = dfc_sc["fixation time"] / normvalue
    normalized_values.append(normalized_value)
df["normalized fixation time"] = pd.concat(normalized_values).tolist()
```
